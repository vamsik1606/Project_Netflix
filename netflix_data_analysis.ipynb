```python
import matplotlib.pyplot as plt

dataframes = {
    "Movies": movies_df,
    "Recommendations": recomm_df,
    "Reviews": reviews_df,
    "Search": search_df,
    "Users": users_df,
    "Watch History": watch_df
}

for name, df in dataframes.items():
    # Count NaN per column
    nan_counts = df.isnull().sum()
    
    # Filter only columns that have at least 1 NaN
    nan_counts = nan_counts[nan_counts > 0]
    
    # If no NaNs, skip plotting
    if nan_counts.empty:
        print(f"✅ {name}: No missing values found.\n")
        continue
    
    # Plot
    plt.figure(figsize=(8, 4))
    nan_counts.plot(kind='bar', color='tomato', edgecolor='black')
    plt.title(f"Missing Values per Column - {name} Dataset")
    plt.ylabel("Number of NaN values")
    plt.xlabel("Columns")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

```python

dataframes = {
    "Movies": movies_df,
    "Recommendations": recomm_df,
    "Reviews": reviews_df,
    "Search": search_df,
    "Users": users_df,
    "Watch History": watch_df
}

for name, df in dataframes.items():
    total_cells = df.shape[0] * df.shape[1]  # total number of cells in the DataFrame
    nan_counts = df.isnull().sum()
    nan_percent = (nan_counts / df.shape[0]) * 100  # percent of rows that have NaN in that column
    
    # Keep only columns with missing values
    nan_percent = nan_percent[nan_percent > 0]
    
    # Skip if there are no NaNs
    if nan_percent.empty:
        print(f"✅ {name}: No missing values found.\n")
        continue

    # Pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(
        nan_percent,
        labels=nan_percent.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.tab10.colors  # gives different colors per column
    )
    plt.title(f"🧩 NaN Percentage per Column - {name} Dataset\n(Total Rows: {df.shape[0]}, Columns: {df.shape[1]})")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

movie_review_user_df['combined_features'] = (
    movie_review_user_df['genre_primary'].fillna('') + ' ' +
    movie_review_user_df['genre_secondary'].fillna('') + ' ' +
    movie_review_user_df['language'].fillna('')
)

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movie_review_user_df['combined_features'])

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

indices = pd.Series(movie_review_user_df.index, index=movie_review_user_df['title']).drop_duplicates()

def recommend_similar_movies(title, n=5):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:n+1]
    movie_indices = [i[0] for i in sim_scores]
    return movie_review_user_df.iloc[movie_indices][['title', 'genre_primary', 'language']]
```
