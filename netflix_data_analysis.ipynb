```python
import pandas as pd
```

```python
# loading the csv files
movies_df = pd.read_csv("dataset/movies.csv")
recomm_df = pd.read_csv("dataset/recommendation_logs.csv")
reviews_df = pd.read_csv("dataset/reviews.csv")
search_df = pd.read_csv("dataset/search_logs.csv")
users_df = pd.read_csv("dataset/users.csv")
watch_df = pd.read_csv("dataset/watch_history.csv")
```

```python
#understand the dataframes
for name, df in {
    "Movies": movies_df,
    "recomendations":recomm_df,
    "reviews":reviews_df,
    "saerch":search_df,
    "Users": users_df,
    "Watch History": watch_df
}.items():
    print(f"{name}:{df.shape}:{df.columns.tolist()}\n")
```

```python
import matplotlib.pyplot as plt

dataframes = {
    "Movies": movies_df,
    "Recommendations": recomm_df,
    "Reviews": reviews_df,
    "Search": search_df,
    "Users": users_df,
    "Watch History": watch_df
}

for name, df in dataframes.items():
    # Count NaN per column
    nan_counts = df.isnull().sum()
    
    # Filter only columns that have at least 1 NaN
    nan_counts = nan_counts[nan_counts > 0]
    
    # If no NaNs, skip plotting
    if nan_counts.empty:
        print(f"✅ {name}: No missing values found.\n")
        continue
    
    # Plot
    plt.figure(figsize=(8, 4))
    nan_counts.plot(kind='bar', color='tomato', edgecolor='black')
    plt.title(f"Missing Values per Column - {name} Dataset")
    plt.ylabel("Number of NaN values")
    plt.xlabel("Columns")
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
